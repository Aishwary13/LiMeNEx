```python
import os
import pandas as pd

# Define the path to the folder containing CSV files
folder_path = 'D:/Raylab/LiMeNEx/FetchData/data/FattyAcidSynthesis'

# Create an empty list to store individual dataframes
df_list = []

# Loop through all the files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        # Construct the full file path
        file_path = os.path.join(folder_path, filename)
        
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path)
        
        # Append the DataFrame to the list
        df_list.append(df)

# Concatenate all dataframes in the list
combined_df = pd.concat(df_list, ignore_index=True)

# Now `combined_df` contains all the data from the CSV files concatenated together
combined_df.head()
```

```python
combined_df = combined_df.drop([list(combined_df.columns)[0], list(combined_df.columns)[1],'Database'], axis = 1)
combined_df['Chea'] = ""
combined_df['Signor'] = ""
combined_df['Trrust'] = ""
```

Signor

```python
Signor_path = 'D:/Raylab/LiMeNEx/FetchData/data/FattyAcidSynthesis/Signor/FattyAcidSynthesis - Signor.csv'
SigDf = pd.read_csv(Signor_path)
SigDf.head()
```

```python
SigDf = SigDf[(SigDf['validated genes'] == 'Yes')]
```

```python
for i in range(0,len(SigDf)):
    tissueList = [var.lower().strip() for var in SigDf.iloc[i]['Tissue'].split(',')]
    TF = SigDf.iloc[i]['TF']
    TargetGene = SigDf.iloc[i]['TargetGene']
    PMID = SigDf.iloc[i]['Library/PMID']

    for tis in tissueList:
        
        mask = (combined_df['TF'] == TF) & (combined_df['TargetGene'] == TargetGene) & (combined_df['Tissue'] == tis)
        temp = combined_df[mask]

        if(len(temp) != 0):
            if(len(temp) > 1):
                print('Here')
            index = temp.index[0]
            combined_df.at[index,'Signor'] += f"{str(PMID)},"
        else:
            newrow = {"TF":TF, 'TargetGene':TargetGene,'Tissue':tis,'Signor':str(PMID)}
            combined_df = pd.concat([combined_df,pd.DataFrame([newrow])], ignore_index=True)
```
